```python
import os
import pandas as pd
import numpy as np
import pathlib
```

```python
data_path = pathlib.Path(os.getcwd()).parent.joinpath("Records")
csv_files = [file for file in os.listdir(data_path) if file.endswith('.csv')]


combined_data = pd.concat([pd.read_csv(data_path.joinpath(file)) for file in csv_files],ignore_index=True)
combined_data.set_index("Unnamed: 0", inplace=True)
combined_data.index.name = "Experiments"

combined_data["Mean_Score"] = combined_data.mean(axis=1)
combined_data.sort_values(by="Mean_Score", inplace=True)
combined_data
```
