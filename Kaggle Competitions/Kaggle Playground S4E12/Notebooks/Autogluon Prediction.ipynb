```python
from autogluon.tabular import TabularPredictor
import pandas as pd
import numpy as np
import os
import pathlib
```

```python
train = pd.read_csv("/kaggle/input/competition/oof_data/oof_data.csv")
test = pd.read_csv("/kaggle/input/competition/oof_data/test_data.csv")
```

```python
MAIN_PATH = "/kaggle/input/competition"
csv_files = [file for file in os.listdir(MAIN_PATH) if file.endswith('.csv')]


combined_data = pd.concat([pd.read_csv(os.path.join(pathlib.Path(MAIN_PATH).joinpath(file))) for file in csv_files],ignore_index=True)
combined_data.set_index("Unnamed: 0", inplace=True)
combined_data.index.name = "Experiments"

combined_data["Mean_Score"] = combined_data.mean(axis=1)
combined_data.sort_values(by="Mean_Score", inplace=True)
combined_data
```

```python
predictor = TabularPredictor(label='Premium_Amount', 
                            eval_metric='rmse',
                            ) 

TIME = 3600*10

predictor.fit(train,
              presets='best_quality',
              time_limit=TIME,
              verbosity=1,
              excluded_model_types=['KNN'],
              ag_args_fit={'num_gpus': 1}
              )

predictor.save("AutoGluon_Model")
results = predictor.fit_summary()
```

```python
predictor.leaderboard()
```

```python
y_pred = predictor.predict(test)
y_pred = np.expm1(y_pred)

sub = pd.read_csv('/kaggle/input/playground-series-s4e12/sample_submission.csv')
sub['Premium Amount'] = y_pred
sub.to_csv('submission.csv', index=False)
```
