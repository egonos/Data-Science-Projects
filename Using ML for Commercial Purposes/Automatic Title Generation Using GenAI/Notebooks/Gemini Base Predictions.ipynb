```python
import google.generativeai as genai
from google.colab import userdata
from time import sleep
import time
import json
import os
import pathlib

from google.colab import drive
drive.mount('/content/drive')
```

```python
#gemini 1.5 flash model for base title generation
base_model = model = genai.GenerativeModel("gemini-1.5-flash")
genai.configure(api_key=userdata.get('GOOGLE_API_KEY'))

#generate 30 titles about cars having maximum 1000 tokens
#Note: I generated 30 titles at once due to gemini quota limits
TOPIC = "cars"
NUMBER_OF_TITLES = 30
config = {
  "max_output_tokens": 1000,
  "temperature": 2,
}


response = model.generate_content(f"You are a helpful ecommerce assistant.\
        Please generate {NUMBER_OF_TITLES} different email titles about {TOPIC} separated by ';' to enhance user engagement.\
        Do not generate any other comments.",
          generation_config = genai.types.GenerationConfig(**config))

base_response = response.text.split(";")

#save base response
with open("/content/drive/MyDrive/SETROW/Data/gemini_base_response.json", "w") as f:
  json.dump(base_response, f)
```
